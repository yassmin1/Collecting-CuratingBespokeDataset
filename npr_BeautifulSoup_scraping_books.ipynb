```python
'''
By ryassminh@y
scraping NPR articles
'''
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
import re
```

```python
# function to extract html document from given url

def check_url(url:str) -> bool:
    reglink =r'[A-Za-z0-9]+://www.npr.org[A-Za-z0-9%-_]+(/[A-Za-z0-9%-_])*(#|\\?)[A-Za-z0-9%-_&=]*'
    p=re.compile(reglink)
    if (url == None):
        return False
    if(re.search(p, url)):
        return True
    else:
        return False

def get_NPR_article(url:str) -> dict:
    """Parse bbc article and return text in list of strings"""
  
    if (len(url.split("/"))==8) & (check_url(url)):
        #print(check_url(url))
        #print(len(url.split("/"))==8)      
        article = requests.get(url)
        soup = BeautifulSoup(article.content, "html.parser")
        #body = soup.find("div",attrs={"id":"storytext"})
        body = soup.find("article",attrs={"class":"story"})
        text=[dd.text.strip() for dd in body.find_all('p')] 
        time= body.find('time')["datetime"]
        header=body.find('h1').get_text(strip=True)
        if  body.find('h3').find('a'):
            news_category=body.find('h3').find('a').text
        else:
            news_category=soup.find('header',attrs={'class':'contentheader contentheader--one'}).find('h2').text
    else:
         text=[]
         time=[]
         header=[]
         news_category=[]
 
    return  {
        'Article':text,
        'DataTime':time,  
        'Header':header,
        'news_category':news_category
        } 

url_input = [
             #'https://www.npr.org/sections/book-reviews',
             'https://www.npr.org/sections/news/' 
             ]

def build_NPR_dataset(seed_urls):
    new_data = []
    for url in seed_urls:
        data = requests.get(url)
        soup = BeautifulSoup(data.content, 'html.parser')
        ff=soup.find_all('h2',attrs={'class':'title'})
        num=len(ff)
        print(f'Number of articles is {num}')
        for i,result in enumerate(ff[:num]):
            if len(result.find('a').get('href')):
                print(f"article num {i} and link {result.find('a').get('href')}")
                article_data=get_NPR_article(result.find('a').get('href'))
                new_data.extend([article_data]) 
            else:
                continue     
            
    df = pd.DataFrame(new_data)
    #output_dir/outputcsvfile
    df.to_csv('npr_books_reviews_articles.csv') 
    return df
   
build_NPR_dataset(url_input)
```
